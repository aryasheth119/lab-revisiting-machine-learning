Checking for null values in all the columns

```python
# Check for null values in all columns
null_values = data.isnull()

# Summarize the count of null values in each column
total_null_values = null_values.sum()

display(total_null_values)
```

Create a new empty list called drop_list. We will append to this list a set of columns to be droped later. Add the following columns to this:

OSOURCE - symbol definitions not provided, too many categories
ZIP - we are including state already

```python
drop_list = []

# Append the specified columns to the drop_list

drop_list.append("OSOURCE")

drop_list.append("ZIP")

display(drop_list)
```

Identify columns that over 85% missing values and add them to the previous list.

```python
nulls_percent_df = pd.DataFrame(data.isna().sum()/len(data)).reset_index()
nulls_percent_df.columns = ['column_name', 'nulls_percentage']
nulls_percent_df
```

```python
nulls_percent_df[nulls_percent_df['nulls_percentage']!=0]
```

```python
columns_above_value = nulls_percent_df[nulls_percent_df['nulls_percentage']>0.85]
columns_above_value['column_name']
```

```python
drop_columns_list = list(columns_above_value['column_name'])
print(drop_columns_list)
```

```python
drop_list.extend(drop_columns_list)

display(drop_list)
```

```python
display(data.columns)
```

Remove the columns included in the drop_list from the dataframe

```python
data.drop(drop_list, axis=1, inplace=True)
display(data)
```

Now, reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

```python
categorical = data.select_dtypes([object])
categorical.head()
```

```python
if 'GENDER' in categorical.columns:
    print("The 'Gender' column exists.")
else:
    print("The 'Gender' column does not exist.")
```

```python
categorical['GENDER'].value_counts()
```

```python
print(categorical['GENDER'].value_counts())
```

```python
categorical['GENDER'] = categorical['GENDER'].fillna('F')
```

```python
df = pd.DataFrame(categorical['GENDER'].value_counts()).reset_index()
df.columns = ['GENDER', 'count']
others = list(df[df['count'] < 2500]['GENDER'])

def clean_gender(x):
    if x == 'M':
        return 'M'
    elif x == 'F':
        return 'F'
    else:
        return 'other'

categorical['GENDER'] = categorical['GENDER'].apply(clean_gender)
```
